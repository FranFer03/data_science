```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('https://curso-data-science-verdadero-6to.s3.sa-east-1.amazonaws.com/csv/videogamesales.csv')
data.head(2)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['Year'] = data['Year'].astype('Int64')
```

```python
data.head()
```

```python
data[ data['Year'].isnull() ]
```

```python
data[ data['Publisher'].isnull() ].head()
```

```python
data.dropna(subset=['Publisher'], inplace=True)
```

```python
data = data.dropna(subset=['Year'])
```

```python
data.isnull().sum()
```

```python
data.drop(columns=['Rank'],inplace=True)
```

```python
data.head()
```

```python
data['Total_Sales'] = data['NA_Sales'] + data['EU_Sales'] + data['JP_Sales'] + data['Other_Sales']
data.head()
```

```python
# Top Publishers by total sales
publishers_by_sales = data.groupby('Publisher')['Total_Sales'].sum()
publishers_by_sales = publishers_by_sales.sort_values(ascending=False)
publishers_by_sales = publishers_by_sales.head(10)
publishers_by_sales
```

```python
publishers_by_sales.plot(kind='bar', figsize=(12, 4), title='Top Publishers by Total Sales')
```

```python
sorted_publishers_by_sales = data.groupby('Publisher')['Total_Sales'].count()
sorted_publishers_by_sales = sorted_publishers_by_sales.sort_values(ascending=False).reset_index()
sorted_publishers_by_sales = sorted_publishers_by_sales.rename(columns={
    'Total_Sales': 'Count'
    }).head(10)
sorted_publishers_by_sales
```

```python
fig,ax = plt.subplots(figsize=(12, 3))
sns.barplot(data=sorted_publishers_by_sales, 
            x='Publisher', 
            y='Count', 
            ax=ax,
            palette='viridis')
ax.set_title('Top Publishers by Number of Games')
plt.xticks(rotation=90)
plt.show()
```

```python
data.head()
```

```python
yearly_sales = data.groupby('Year').agg(
    Total_Sales=('Total_Sales', 'sum'),
    Count_Games=('Total_Sales', 'count')
).reset_index()
yearly_sales = yearly_sales.sort_values(by='Year')
yearly_sales.head()
```

```python
yearly_sales.plot(x='Year', y='Total_Sales', kind='line', figsize=(12, 4), title='Yearly Total Sales of Video Games')
```

```python
yearly_sales['Performance'] = round(yearly_sales['Total_Sales']/yearly_sales['Count_Games'],2)
yearly_sales[ yearly_sales['Year'] >= 2015]
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
sns.lineplot(data=yearly_sales, 
             x='Year', 
             y='Total_Sales', 
             ax=ax, 
             marker='o', 
             color='orange')
sns.lineplot(data=yearly_sales,
             x='Year', 
             y='Count_Games', 
             ax=ax, 
             marker='o', 
             color='blue')
ax.set_title('Yearly Total Sales and Number of Games')
```
